```python
import pandas as pd
import matplotlib.pyplot as plt

class SalesAnalysis:

    def __init__(self, file_name):
        self.file_name = file_name
        self.df = None

    
    def load_data(self):
        self.df = pd.read_csv(self.file_name)
        print("Dataset Loaded Successfully\n")

    
    def show_data(self):
        print(self.df.head())


    def total_sales(self):
        total = self.df["Sales"].sum()
        print("\nTotal Sales:", total)


    def top_products(self):
        top = self.df.groupby("Product")["Sales"].sum()

        print("\nTop Selling Products:")
        print(top)

        top.plot(kind="bar")

        plt.title("Sales by Product")
        plt.xlabel("Product")
        plt.ylabel("Sales")

        plt.show()
        
obj = SalesAnalysis("sales_data.csv")

obj.load_data()
obj.show_data()
obj.total_sales()
obj.top_products()
```
